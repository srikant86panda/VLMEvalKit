```python
import os
import re
import glob
import pandas as pd

def get_group(filename):
    pattern = r'grid_\d+x\d+'
    match = re.search(pattern, filename)
    if match:
        return match.group()
    return 'orig'

def get_patch(filename):
    pattern = r'row\d+_col\d+'
    match = re.search(pattern, filename)
    if match:
        return match.group()
    return 'orig'

base_dir = './outputs'
dataset_name = 'BLINK_sample'

overall_results = {}

model_names = ['llava-onevision-qwen2-0.5b-ov-hf', 'InternVL2-1B', 'InternVL2_5-1B', 'Phi-3-Vision']

for model_name in model_names:
    file_paths = glob.glob(f'{base_dir}/{model_name}/{model_name}_{dataset_name}_*openai_result.xlsx')
    groups = {}

    for file_path in file_paths:
        group = get_group(os.path.basename(file_path))
        if group not in groups:
            groups[group] = [file_path]
        else:
            groups[group].append(file_path)

    for key, values in groups.items():
        if key == 'orig':
            continue  # Skip the 'orig' key

        print(f"\nProcessing group: {key}")
        dataframes = {os.path.basename(f): pd.read_excel(f) for f in values}

        # Ensure all dataframes have the same shape and 'hit' column
        for df in dataframes.values():
            if 'hit' not in df.columns:
                raise ValueError(f"Missing 'hit' column in one of the files in group {key}.")

        # Initialize results for the group
        group_results = {}

        for file1, df1 in dataframes.items():
            # Logical condition for file1 having hit=1 while others have hit=0
            other_files = {f: df for f, df in dataframes.items() if f != file1}

            count_1_only = 0

            for idx in range(len(df1)):
                if df1.loc[idx, "hit"] == 1 and all(
                    df.loc[idx, "hit"] == 0 for df in other_files.values()
                ):
                    count_1_only += 1

            # Build configuration name
            patch = get_patch(file1)
            config = f"{key}_{patch}"
            if config not in group_results:
                group_results[config] = {}
            group_results[config][model_name] = count_1_only

        # Merge group results into overall results
        for config, counts in group_results.items():
            if config not in overall_results:
                overall_results[config] = {}
            overall_results[config].update(counts)

# Convert overall results to DataFrame
overall_results_df = pd.DataFrame.from_dict(overall_results, orient='index').reset_index()
overall_results_df.columns = ['Configuration'] + model_names

# Fill missing values with 0
overall_results_df = overall_results_df.fillna(0)

# Display final DataFrame)
```

```python
overall_results_df.to_csv("blink_test.csv",index=False)
```
